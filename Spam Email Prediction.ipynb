Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer # text used multiple times would get a values
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading data into a pandas data frame
mail_data = pd.read_csv("F:\Projects\ML projects\Copy project\Spam Email Prediction\mail_data.csv")
```

```python
mail_data.head()
```

```python
# Checking for null values
mail_data.isnull().sum()
```

```python
mail_data.shape
```

```python
# labelling spam as 0 and ham as 1
mail_data.loc[mail_data['Category'] == 'spam','Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham','Category'] = 1
```

```python
x = mail_data['Message']
y = mail_data['Category']
```

```python
# Splitting the data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x, x_train, x_test)
```

Feature Extraction

```python
feature_extraction = TfidfVectorizer( min_df= 1, stop_words='english', lowercase = True)
```

```python
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features =  feature_extraction.transform(x_test)
```

```python
# converting y data into integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train_features)
```

Training the Data using Logistic Regression Model

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

```python
# predicting the data
x_train_predict = model.predict(x_train_features)
```

```python
# Accuracy of training data
accuracy_training_data = accuracy_score(y_train, x_train_predict)
```

```python
print('Accuracy on training Data : ', accuracy_training_data)
```

```python
# predicting for test data
x_test_predict = model.predict(x_test_features)
# Accuracy of test data
accuracy_testing_data = accuracy_score(y_test, x_test_predict)

print('Accuracy on test Data : ', accuracy_testing_data)
```
